```python
reset()
#a
t,w0=var("t,w0")
x=function("x")(t)
assume(w0>0)
eq=diff(x,t,2)+w0^2*x==0
sola=desolve(eq,[x,t])
show(sola)
```

```python
#b
x0,v0=var("x0,v0")
solb(t,w0,x0,v0)=desolve(eq,[x,t],[0,x0,v0])
show(solb)
#c
solc=solb(t,sqrt(9.81/0.3924),0.15,0)
show(solc)
plot(solc,t,0,3)
```

```python
reset()
#a
t,w0,w,F0=var("t,w0,w,F0")
x=function("x")(t)
assume(w0>0)
assume(w>0)
assume(w0!=w)
eq=diff(x,t,2)+w0^2*x==F0*cos(w*t)
sola=desolve(eq,[x,t])
show(sola)
#b
x1,x2=var("x1,x2")
sol(t,w0,w,F0,x1,x2)=desolve(eq,[x,t],[0,x1,x2])
solb=sol(t,5,5.5,2,0,0)
show(solb)
plot(solb,t,0,50)
```
