```python
# 1
# a)
reset()
x=var("x")
y=function("y")(x)
eq2=(x^2 -1)*diff(y,x)+2*x*y^2 == 0
_C=var("_C")
sol=desolve(eq2,y)
sol1(x,_C)=solve(sol,y)[0].rhs()
show(sol1(x,_C))
list_f=[sol1(x,i/10) for i in [10..20]]
plot(list_f,x,2,5)
```

```python
# 1
# c)
reset()
x=var("x")
y=function("y")(x)
eq2=2*x^2*diff(y,x)==x^2+y^2
_C=var("_C")
sol=desolve(eq2,y)
show(sol)
sol1(x,_C)=solve(sol,y)[1].rhs()
show(sol1(x,_C))
list_f=[sol1(x,i/10) for i in [10..20]]
plot(list_f,x,0,5,ymin=-10,ymax=10,detect_poles="True")
```

```python
# 1
# d)
reset()
x=var("x")
y=function("y")(x)
eq2=diff(y,x)==-(x+y)/y
_C=var("_C")
sol=desolve(eq2,y)
show(sol)
yy= var("yy")
f(x,yy,_C)=sol.substitute(y(x)==yy)
show(f(x,yy,_C))
g=implicit_plot(f(x,yy,-4),(x,-10,10),(yy,-10,10))
for i in [-20..20]:
    g1=implicit_plot(f(x,yy,i/5),(x,-10,10),(yy,-10,10))
    g=g+g1
show(g)
```
