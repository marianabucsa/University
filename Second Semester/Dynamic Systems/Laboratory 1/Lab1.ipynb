```python
t=var("t")
f(t,s)=1-(s*cos(4*t)*cos(t))/sqrt(1-s^2*(cos(4*t))^2*(sin(t))^2)
x(t,s)=f(t-pi/2,s)
y(t,s)=f(t,s)
g=parametric_plot((x(t,0.1),y(t,0.1)),(t,0,2*pi))
for k in [2 .. 10]:
    g1=parametric_plot((x(t,k),y(t,k)),(t,0,2*pi))
    g=g+g1
show(g)
```
